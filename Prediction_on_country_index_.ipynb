3 Write Python code that will print the names of all the countries that are outliers in term of the density of population per square km3 (Density pop./km2 column).

After executing the code, it shows that Countries with Outliers in Density pop./km2:

33    Singapore

66    Hong Kong

```python
import pandas as pd

# Assuming 'default' is your DataFrame
# Replace 'your_file_path.csv' with the actual file path if needed
# default = pd.read_csv('your_file_path.csv')

# Display the first few rows of the DataFrame
print("Preview of the DataFrame:")
print(default.head())

# Calculate Z-scores for the 'Density pop./km2' column
z_scores = (default['Density pop./km2'] - default['Density pop./km2'].mean()) / default['Density pop./km2'].std()

# Define a threshold for outliers (e.g., Z-score greater than 3)
outlier_threshold = 3

# Identify countries with Z-scores greater than the threshold
outliers = default[z_scores.abs() > outlier_threshold]

# Print the names of countries that are outliers
print("\nCountries with Outliers in Density pop./km2:")
print(outliers['Country'])
```

```python
import pandas as pd

# Assuming 'default' is your DataFrame
# Replace 'your_file_path.csv' with the actual file path if needed
# default = pd.read_csv('your_file_path.csv')

# Select relevant columns for analysis
columns_of_interest = ["Quality of Life Index", "Purchasing Power Index", "Pollution Index",
                        "Safety Index", "Cost of Living Index", "Restaurant Price Index", "Climate Index"]

# Subset the DataFrame with the selected columns
subset_data = default[columns_of_interest]

# Calculate the correlation matrix
correlation_matrix = subset_data.corr()

# Find the two columns with the strongest positive correlation
strongest_positive_correlation = correlation_matrix.unstack().sort_values(ascending=False).drop_duplicates()

# Extract column names with the strongest positive correlation
column1, column2 = strongest_positive_correlation.index[1]

# Print the result
print(f"The two columns with the strongest positive correlation are '{column1}' and '{column2}' with a correlation of {strongest_positive_correlation[1]:.2f}")
```

5.Write a code to produce graphs in Python to visualise the following information:

(I) How many countries are in the following bands based on population: small (less than 10,000,000), medium (10,000,000 â€“ 100,000,000) and large (100,000,000+)


```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'default' is your DataFrame
# Replace 'your_file_path.csv' with the actual file path if needed
# default = pd.read_csv('your_file_path.csv')

# Display the first few rows of the DataFrame
print("Preview of the DataFrame:")
print(default.head())

# Define population bands
small_band = 10000000
medium_band = 100000000

# Create a new column 'Population Band' based on population
default['Population Band'] = pd.cut(default['Population'], bins=[0, small_band, medium_band, float('inf')],
                                   labels=['Small (<10M)', 'Medium (10M-100M)', 'Large (100M+)'])

# Count the number of countries in each population band
band_counts = default['Population Band'].value_counts()

# Plot the bar chart
plt.figure(figsize=(8, 6))
band_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Countries in Population Bands')
plt.xlabel('Population Band')
plt.ylabel('Number of Countries')
plt.show()
```

Explanation

From the graph above, its shows that most countries with medium population are more in number followed by small number and lowest with large population

5. ii. What percentage of population was obtained using each of the possible population sources (Official quarterly estimate, official provisional figure, Official estimateâ€¦)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'default' is your DataFrame
# Replace 'your_file_path.csv' with the actual file path if needed
# default = pd.read_csv('your_file_path.csv')

# Display the first few rows of the DataFrame
print("Preview of the DataFrame:")
print(default.head())

# Group by 'Population source' and calculate the percentage of population for each source
population_percentage = default.groupby('Population source')['Population'].sum() / default['Population'].sum() * 100

# Plot the pie chart
plt.figure(figsize=(8, 8))

# Dynamically set explode based on the number of unique categories
explode_values = [0.1] * len(population_percentage)

population_percentage.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightcoral', 'lightgreen'],
                            wedgeprops=dict(width=0.3), explode=explode_values, shadow=True)

plt.title('Percentage of Population by Population Source')
plt.ylabel('')
plt.show()
```
